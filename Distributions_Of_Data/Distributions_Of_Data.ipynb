```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
```

Data Creation for the income

```python
np.random.seed(0)
income_data = np.random.gamma(shape=2.0, scale=10000.0, size=100000)
```

```python
skew_value = skew(income_data)

sns.histplot(income_data, bins=30, kde=True)
plt.title(f"Incom Data (skewness: {skew_value:.2f})")
plt.xlabel('Incom')
plt.ylabel('Frequency')
```
